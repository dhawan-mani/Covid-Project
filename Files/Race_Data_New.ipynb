```python
%matplotlib inline
import pandas as pd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
data = pd.read_csv("Race_Data_Updated.csv")
df = pd.DataFrame(data)
```

```python

df.fillna(0,inplace=True)
type(df["Cases_White"])
df['Cases_White'] = pd.to_numeric(df['Cases_White'], errors='coerce')
df['Cases_Total'] = pd.to_numeric(df['Cases_Total'], errors='coerce')
New_df =df.head(56)
New_df
```

```python
# Sum_Data.reset_index()
BlackAfricanAmerican = New_df["Cases_Black"].sum()
Latin = New_df["Cases_LatinX"].sum()
Asians = New_df["Cases_Asian"].sum()
AmericanIndianAlaskaNative = New_df["Cases_AIAN"].sum()
Multiracial = New_df["Cases_Multiracial"].sum()
Hispanic  = New_df["Cases_Ethnicity_Hispanic"].sum()
NativeHawaiianPacificIslander = New_df["Cases_NHPI"].sum()
Other = New_df["Cases_Other"].sum()
Unknown = New_df["Cases_Unknown"].sum()
White = New_df["Cases_White"].sum()

Dict = {"Black African American" :BlackAfricanAmerican,
        "Asians": Asians,
        "American Indian Alaska Native":AmericanIndianAlaskaNative,
        "Multiracial": Multiracial,
        "Hispanic or latino" :Hispanic,
        "Native Hawaiian_Pacific Islander":NativeHawaiianPacificIslander,
        "Other" :Other,
        "White":White,
       
    
}
Races_df = pd.DataFrame([Dict])
Races_df 
```

```python
Races_df_t = Races_df.transpose()
# Races_df.plot(kind="bar")
Races_df_t.reset_index(inplace=True)
Cases_DF = Races_df_t.rename(columns={"index":"Race",0:"Total Cases"})
Cases_DF
```

```python
colors = ["red","blue","green","yellow","black","orange","violet","purple","pink","brown"]
ax = Cases_DF.plot(kind='barh',x='Race',y='Total Cases',stacked=True,color=colors)
plt.legend(fancybox=True,shadow=True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.savefig("RACE DATA IMAGES/TotalCases.png")
```

### Analysis:
### Whites have the highest no. of cases reported in the U.S.

```python
BlackAfricanAmerican = New_df["Deaths_Black"].sum()
Latin = New_df["Deaths_LatinX"].sum()
Asians = New_df["Deaths_Asian"].sum()
AmericanIndianAlaskaNative = New_df["Deaths_AIAN"].sum()
Multiracial = New_df["Deaths_Multiracial"].sum()
Hispanic  = New_df["Deaths_Ethnicity_Hispanic"].sum()
NativeHawaiianPacificIslander = New_df["Deaths_NHPI"].sum()
Other = New_df["Deaths_Other"].sum()
White = New_df["Deaths_White"].sum()
Unknown = New_df["Cases_Unknown"].sum()

Death_dict = {"Black African American" :BlackAfricanAmerican,
        "Asians": Asians,
        "American Indian Alaska Native":AmericanIndianAlaskaNative,
        "Multiracial": Multiracial,
        "Hispanic or latino" :Hispanic,
        "Native Hawaiian_Pacific Islander":NativeHawaiianPacificIslander,
        "Other" :Other,
        "White":White,
       
    
}
Deaths_df = pd.DataFrame([Death_dict])
Deaths_df 
```

```python
Deaths_df_t = Deaths_df.transpose()

Deaths_df_t.reset_index(inplace=True)
Deaths_df_t = Deaths_df_t.rename(columns={"index":"Race",0:"Total Deaths"})
Deaths_df_t
```

```python
from numpy.random import rand
colors = ["red","blue","green","yellow","black","orange","violet","purple","pink","brown"]
axis = Deaths_df_t.plot(kind='barh',x='Race',y='Total Deaths',stacked=True,color=colors)
plt.legend(fancybox=True,shadow=True)
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.spines['bottom'].set_visible(False)
axis.spines['left'].set_visible(False)
plt.savefig("RACE DATA IMAGES/TotalDeaths.png")
```

### Analysis:

Same trend can be seen for Hispanic or Latino and Black African Americans too.

```python
Percentage_Data = pd.merge(Cases_DF,Deaths_df_t,how="outer",on="Race")
Percentage_Data
```

```python
#Reference for population:https://data.census.gov/cedsci/table?q=United%20States&g=0100000US&tid=ACSDP1Y2018.DP05
Pop = {"Black African American":41617764,
       "Asians":18415198,
       "American Indian Alaska Native":2801587,
       "Multiracial":11280031,
        "Hispanic or Latino":59763631,
       "Native Hawaiian_Pacific Islander":626054,
       "Other":16253785,
       "White":236173020

       }
Percentage_Data["Population"] =" "
population_rate=[]

for key,values in Pop.items():
        population_rate.append(values)
print(population_rate)
for index,row in Percentage_Data.iterrows():
    row["Population"] = population_rate[index]
    percent = round(row["Total Deaths"]/row["Population"]*100000,0)
    Percentage_Data.loc[index,"Population"] = row["Population"]
    Percentage_Data.loc[index,"Percentage Death"] = percent
```

```python
Percentage_Data
```

```python
from numpy.random import rand
axis = Percentage_Data.plot(kind='barh',x='Race',y='Percentage Death',stacked=True,cmap="Paired")
y_value = list(Percentage_Data["Percentage Death"])
#Reference for adding values above the bar :https://www.kite.com/python/answers/how-to-display-the-value-of-each-bar-in-a-bar-chart-using-matplotlib-in-python
for index, value in enumerate(y_value):           
    axis.text(value, index, str(value))
plt.legend(fancybox=True,shadow=True)
plt.title("Deaths per 100,000 people by race or ethnicity",fontsize=20,fontweight=10,color="red")
plt.xlabel("Nationwide, Black people are dying at 2.5 times the rate of white people.",fontsize=15,fontweight=25)
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.spines['bottom'].set_visible(False)
axis.spines['left'].set_visible(False)
plt.savefig("RACE DATA IMAGES/Percentage_Death_Chart.png")    
```
