```python
%matplotlib inline
import pandas as pd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
data = pd.read_csv("Race_Data_Updated.csv")
df = pd.DataFrame(data)
```

```python

df.fillna(0,inplace=True)
type(df["Cases_White"])
df['Cases_White'] = pd.to_numeric(df['Cases_White'], errors='coerce')
df['Cases_Total'] = pd.to_numeric(df['Cases_Total'], errors='coerce')
New_df = df.head(56)
New_df
```

```python
Ethnicity_Data = New_df.groupby("State")
Sum_Data = Ethnicity_Data.sum()

#Ethnicity_Data.to_csv("Output.csv")
Sum_Data.reset_index(inplace=True)
del Sum_Data["Date"]
del Sum_Data["State"]
#Sum_Data.reset_index(inplace=True)
Sum_Data
```

```python
# Sum_Data.reset_index()
BlackAfricanAmerican = Sum_Data["Cases_Black"].sum()
Latino = Sum_Data["Cases_LatinX"].sum()
Asians = Sum_Data["Cases_Asian"].sum()
AmericanIndianAlaskaNative = Sum_Data["Cases_AIAN"].sum()
Multiracial = Sum_Data["Cases_Multiracial"].sum()
Hispanic  = Sum_Data["Cases_Ethnicity_Hispanic"].sum()
NativeHawaiianPacificIslander = Sum_Data["Cases_NHPI"].sum()
Other = Sum_Data["Cases_Other"].sum()
Latins = Sum_Data["Cases_Unknown"].sum()
White = Sum_Data["Cases_White"].sum()
Dict = {"Black African American" :BlackAfricanAmerican,
        "Latino" : Latino,
        "Asians": Asians,
        "American Indian Alaska Native":AmericanIndianAlaskaNative,
        "Multiracial": Multiracial,
        "Hispanic" :Hispanic,
        "Native Hawaiian_Pacific Islander":NativeHawaiianPacificIslander,
        "Other" :Other,
        "Latins":Latins,
        "White":White,
    
}
Races_df = pd.DataFrame([Dict])
Races_df 
```

```python
Races_df_t = Races_df.transpose()
# Races_df.plot(kind="bar")
Races_df_t.reset_index(inplace=True)
DF = Races_df_t.rename(columns={"index":"Race",0:"Total Cases"})
DF
```

```python
colors = ["red","blue","green","yellow","black","orange","violet","purple","pink","brown"]
ax = DF.plot(kind='barh',x='Race',y='Total Cases',stacked=True,color=colors)
plt.legend(fancybox=True,shadow=True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.savefig("RACE DATA IMAGES/TotalCases.png")
```

```python
BlackAfricanAmerican = Sum_Data["Deaths_Black"].sum()
Latino = Sum_Data["Deaths_LatinX"].sum()
Asians = Sum_Data["Deaths_Asian"].sum()
AmericanIndianAlaskaNative = Sum_Data["Deaths_AIAN"].sum()
Multiracial = Sum_Data["Deaths_Multiracial"].sum()
Hispanic  = Sum_Data["Deaths_Ethnicity_Hispanic"].sum()
NativeHawaiianPacificIslander = Sum_Data["Deaths_NHPI"].sum()
Other = Sum_Data["Deaths_Other"].sum()
Latins = Sum_Data["Deaths_Unknown"].sum()
White = Sum_Data["Deaths_White"].sum()
Death_dict = {"Black African American" :BlackAfricanAmerican,
        "Latino" : Latino,
        "Asians": Asians,
        "American Indian Alaska Native":AmericanIndianAlaskaNative,
        "Multiracial": Multiracial,
        "Hispanic" :Hispanic,
        "Native Hawaiian_Pacific Islander":NativeHawaiianPacificIslander,
        "Other" :Other,
        "Latins":Latins,
        "White":White,
    
}
Deaths_df = pd.DataFrame([Death_dict])
Deaths_df 
```

```python
Deaths_df_t = Deaths_df.transpose()

Deaths_df_t.reset_index(inplace=True)
Deaths_df_t = Deaths_df_t.rename(columns={"index":"Race",0:"Total Deaths"})
Deaths_df_t
```
